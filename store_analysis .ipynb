```python
pascucci_read = pd.read_csv('./caffe_pascucci.csv')
pascucci_read
```

```python
starbucks_read = pd.read_csv('./starbucks.csv')
starbucks_read
```

```python
starbucks_sido_index = ['서울특별시', '경기도', '광주광역시', '대구광역시', '부산광역시', '울산광역시', '인천광역시', '강원도', '경상남도', '경상북도', '전라남도', '전라북도', '충청남도', '충청북도','제주특별자치도', '세종특별자치시']
pascucci_sido_index = ['서울', '경기도', '광주', '대구', '부산', '울산', '인천', '강원도', '경상남도', '경상북도', '전라남도', '전라북도', '충청남도', '충청북도', '제주도', '세종시특별자치시']
```

```python
starbucks_서울 = starbucks_read[starbucks_read['sido'] == starbucks_sido_index[0]]
starbucks_서울
```

```python
pascucci_서울 = pascucci_read[pascucci_read['sido'] == pascucci_sido_index[0]]
pascucci_서울
```

```python
print(len(starbucks_서울))
print(len(pascucci_서울))
```

```python
서울_list = []
서울_list.append([len(starbucks_서울)] + [len(pascucci_서울)])
data = pd.DataFrame(서울_list, columns = ['starbucks','pascucci'], index = ['서울'])
data
```

```python
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams['axes.grid'] = True
plt.rc('font', size=15) 

data_bar = data.plot.bar(color=['g','r'], width=0.3)
plt.show()
```

```python
sido_list = []
for i in range(len(pascucci_sido_index)):
    starbucks = starbucks_read[starbucks_read['sido'] == starbucks_sido_index[i]]
    pascucci = pascucci_read[pascucci_read['sido'] == pascucci_sido_index[i]]
    sido_list.append([len(starbucks)] + [len(pascucci)])
sido_data = pd.DataFrame(sido_list, columns = ['starbucks','pascucci'], index = pascucci_sido_index)
sido_data
```

```python
plt.rcParams["figure.figsize"] = (16, 10)
plt.rcParams['axes.grid'] = True
plt.rc('font', size=15) 


sido_data_bar = sido_data.plot.bar(color=['g','r'], width=0.8)


plt.show()
```
